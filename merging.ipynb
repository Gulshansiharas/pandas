```python
import pandas as pd
```

```python
df=pd.DataFrame({'key1':[1,2,4,5,6],'key2':[4,5,6,7,8],'key3':[3,4,5,6,6]})
```

```python
df
```

```python
df1=pd.DataFrame({'key1':[1,2,45,6,7],'key4':[56,5,6,7,8],'key5':[3,56,5,6,6]})
```

```python
df1
```

```python
#To merging the two data
pd.merge(df,df1)
```

```python
#To merge leftside
pd.merge(df,df1,how='left')
```

```python
pd.merge(df,df1,how='right')
```

```python
pd.merge(df,df1,how='cross')
```
